```python
#Problem 2: 

#Creating a class named Employee.
class Employee():
    
    # init function to take argunments.
    def __init__(self, employee_id, employes_name ,employee_salary):
        self.__employee_id=employee_id
        self.__employes_name=employes_name
        self.__employee_salary=employee_salary
    
    #Function to show employees details.
    def emmployee_details(self):
        return f'Employee ID = {self.__employee_id}, Name = {self.__employes_name}, Salary = {self.__employee_salary} INR'
    
    #Function to calculate bonus.
    def calculate_bonus(self):
         bonus_percentage=10
         bonus = self.__employee_salary* (bonus_percentage / 100)
         return f'Your Yearly bonus is {bonus} INR'

#Adding some employee details. 
employee1=Employee(5156,'Aman',350000) 
employee2=Employee(5465,'Vijay',450000)
employee3=Employee(5162,'Ram',300000)
employee4=Employee(8752,'Monu',250000)
employee5=Employee(7623,'Sumit',1200000)

#Printing employee details with 10% of yearly bonus.
print(employee5.emmployee_details(),'\n',employee5.calculate_bonus())
```

```python
#Problem 3: 

#Declaring a class name vechicals.
class Vehical:
    
    # init function to pass car id, name, model and rent rate.
    def __init__(self, car_id, car_name, car_model, rent_rate, avaliable=True):
        self.car_id=car_id
        self.car_name=car_name
        self.car_model=car_model
        self.rent_rate=rent_rate
        self.avaliable=avaliable
    
    #Function to return car details.    
    def __str__(self):
        return f"{self.car_name} {self.car_model} (ID: {self.car_id})"
    
#Declaring a class rent_system for rent and returning cars.   
class rent_system(): 
    
    # init function to store car data.  
    def __init__(self):
        self.cars = []
        
    # Function for adding car in list
    def add_car(self, car):
        self.cars.append(car)   
        
    # Function for checking car is avaliable or not.   
    def aval_vehicals(self):
        avaliable_vehical=[car for car in self.cars if car.avaliable]
        if avaliable_vehical:
            print("Avaliable Cars")
            for car in avaliable_vehical:
                print(f' - {car}')
        else:
            print("No car avaliable")
    
    #Function to rent a vehical.
    def rent_vehical(self, car_id, days):
        for car in self.cars:
            if car.car_id==car_id and car.avaliable:
                rent_cost=car.rent_rate*days
                print(f"Rented {car} for {days} days. Total cost: ${rent_cost}")
                car.available = False
                return   
        print(f"Car with ID {car_id} not found or not available for rent.")    
    
    #Function to return cars.            
    def return_car(self, car_id):
        for car in self.cars:
            if car.car_id == car_id and not car.available:
                car.available = True
                print(f"Returned {car}.")
                return
        print(f"Car with ID {car_id} not found or not currently rented.")
       
# Decaling a variable for rent_system class.               
Car_system=rent_system()  

#Passing car details to the vehical class.     
car1 = Vehical(1, "Toyota", "Camry",  500)
car2 = Vehical(2, "Honda", "Accord", 650)
car3 = Vehical(3, "Ford", "Mustang", 850)    

#Passing car details to add_car function to append in list.
Car_system.add_car(car1)    
Car_system.add_car(car2) 
Car_system.add_car(car3)   

#Testing for renting different cars and returning cars.    
Car_system.aval_vehicals()
Car_system.rent_vehical(3,5)
Car_system.aval_vehicals()
Car_system.rent_vehical(2,3)
Car_system.return_car(3)

```

```python
#Problem 6:

#Declaring class Rectangle for find perimeter, area and volume of rectangle.
class Rectangle():
    
    # init function to take user input.
    def __init__(self,width, length, height):
        self.width=width
        self.length=length
        self.height=height
    
    # Function to calculate volume    
    def volume(self):
        return f'Volume of rectangle: {self.width*self.length*self.height} Cubic Units.'
    
    # Function to calculate area
    def area(self):
        return f'Area of rectangle: {self.width*self.length} Sq. Units.'
    
    # Function to calculate perimeter
    def perimeter(self):
        return f'Perimeter of rectangle: {2*(self.width+self.length)} Units.'

#Declaring class Square for find perimeter, area and volume of square.   
class Square():
    
    def __init__(self,side):
        self.side=side
    
    # Function to calculate volume
    def volume(self):
        return f'Volume of square: {self.side**3} Cubic Units.'
    
    # Function to calculate area
    def area(self):
        return f'Area of square: {self.side**2} Sq. Units.'
    
    # Function to calculate perimeter
    def perimeter(self):
        return f'Perimeter of square: {self.side*4} Units.'
 
#calculating volume of rectangle.  
find=Rectangle(5,3,7)
find.volume()

```

```python
#Problem 7:

class stu_details():
    
    def __init__(self, Student_id, Studen_name, subject1,subject2,subject3, grade=''):
        self.Student_id=Student_id
        self.Studen_name=Studen_name
        self.subject1=subject1
        self.subject2=subject2
        self.subject3=subject3
        self.grade=grade
    
    #Function to return details of student.    
    def student_details(self):
        return f" Student Id: {self.Student_id} \n Name: {self.Studen_name}\n"
        
    #Function to calculate the grade of student based on mark obtain in subjects.
    def Avg_grade(self):
        mark_obtain=self.subject1+self.subject2+self.subject3
        
        if mark_obtain>150:
            return f'Each subject contains maximum 50 marks (3x50=150), {mark_obtain}.'
        elif mark_obtain<0:
            return f'Marks can not be negetive, {self.subject1}, {self.subject2}, {self.subject3}.'
        
        elif mark_obtain>140 and mark_obtain<=150:
            self.grade='A+'
            return self.grade
        elif mark_obtain>120 and mark_obtain<=140:
            self.grade='A'
            return self.grade
        elif mark_obtain>100 and mark_obtain<=120:
            self.grade='B+'
            return self.grade
        elif mark_obtain>80 and mark_obtain<=100:
            self.grade='B'
            return self.grade
        elif mark_obtain>70 and mark_obtain<=80:
            self.grade='C'
            return self.grade
        else:
            self.grade='D'
            return self.grade
        
#Passing student details and marks.        
student1=stu_details(1,"Frank",45,34,37)
student2=stu_details(2,"Bob",44,47,37)
student3=stu_details(3,"Joy",45, 40,27)
student4=stu_details(4,"Mark",45,34,47)
student4=stu_details(5,"Harry",45,49,47)

#Creating a list of student.
Student_list=[student1,student2,student3,student4]

# itrating through the list and printing each student details and obtained grade.
for student in Student_list:
    Stu_detail=student.student_details()
    Stu_grade=student.Avg_grade()
    print(f'\nStudent Details:\n{Stu_detail} Grade: {Stu_grade}')
          
```

```python
# Problem 10:

#Class defination.
class To_Do_List():
    
    def __init__(self, task_list):
        self.task_list=task_list
        
    #Function to display tasks.
    def display_task(self):
         return self.task_list
    
    #Functio to add more task.    
    def add_task(self,add_more_task):
       
        self.task_list.append(add_more_task)
        return  f'Task "{add_more_task}" added to list'
    
    #Function to mark task as complete.
    def mark_task(self, task_number):
        if 1 <= task_number <= len(self.task_list):
            task_index = task_number - 1
            task = self.task_list[task_index]
            self.task_list[task_index] = f"[Completed] {task}"
            return f"Task '{task}' marked as complete."
        else:
            return "Invalid task number."
    
    #Function to delete the task from list.    
    def delete_task(self,task_number):
        if 1 <= task_number <= len(self.task_list):
            task_index = task_number - 1
            removed_task = self.task_list.pop(task_index)
            return f"Task '{removed_task}' removed."
        else:
            return "Invalid task number."

#Adding some taks.
task1='Creating a presentation for project meeting before 01:00 PM.' 
task2='Attend a meeting with client at 2:30 PM.'
task3='Handover your car for serviceing at 04:00 PM.'
task4='Going to take you car from car servicing point at 06:45 PM.'
task5='Going to dinner with your friend at 08:00 PM.'
Tasks=[task1,task2,task3,task4,task5]

#Passing required arrgunment to the class.
to_do_list=To_Do_List(Tasks)

#Function that display a user menu and take user input as menu instructions.
def menu_list(user_input):
    
    #Condition on various user inpute.
    if user_input==1:
       print("\nYour tasks:\n")
       todays_task=to_do_list.display_task()
       for task in todays_task:
           print(task)
    
    elif user_input==2:
       new_task=input("Enter the task you want to add:")
       more_task=to_do_list.add_task(new_task)
       print(more_task) 
         
    elif user_input == 3:
        task_number = int(input("Enter the task number to mark as complete: "))
        result = to_do_list.mark_task(task_number)
        print(result)
        
        
    elif user_input == 4:
        task_number = int(input("Enter the task number to remove: "))
        result = to_do_list.delete_task(task_number)
        print(result)               
    else:
        return None

#Printing the menu.
print('To DO LIST:-\nEnter 1 for display your todays task.\nEnter 2 for adding more task.\nEnter 3 for mark task as complete.\nEnter 4 to delete any task from list.')

#Taking user input and passing it menu function.
choice=int(input("Enter you choice:"))
user=menu_list(choice)
   
```
