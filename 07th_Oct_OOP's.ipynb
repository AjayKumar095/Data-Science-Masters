```python
#Problem 2: 

#Creating a class named Employee.
class Employee():
    
    # init function to take argunments.
    def __init__(self, employee_id, employes_name ,employee_salary):
        self.__employee_id=employee_id
        self.__employes_name=employes_name
        self.__employee_salary=employee_salary
    
    #Function to show employees details.
    def emmployee_details(self):
        return f'Employee ID = {self.__employee_id}, Name = {self.__employes_name}, Salary = {self.__employee_salary} INR'
    
    #Function to calculate bonus.
    def calculate_bonus(self):
         bonus_percentage=10
         bonus = self.__employee_salary* (bonus_percentage / 100)
         return f'Your Yearly bonus is {bonus} INR'

#Adding some employee details. 
employee1=Employee(5156,'Aman',350000) 
employee2=Employee(5465,'Vijay',450000)
employee3=Employee(5162,'Ram',300000)
employee4=Employee(8752,'Monu',250000)
employee5=Employee(7623,'Sumit',1200000)

#Printing employee details with 10% of yearly bonus.
print(employee5.emmployee_details(),'\n',employee5.calculate_bonus())
```

```python
#Problem 3: 

#Declaring a class name vechicals.
class Vehical:
    
    # init function to pass car id, name, model and rent rate.
    def __init__(self, car_id, car_name, car_model, rent_rate, avaliable=True):
        self.car_id=car_id
        self.car_name=car_name
        self.car_model=car_model
        self.rent_rate=rent_rate
        self.avaliable=avaliable
    
    #Function to return car details.    
    def __str__(self):
        return f"{self.car_name} {self.car_model} (ID: {self.car_id})"
    
#Declaring a class rent_system for rent and returning cars.   
class rent_system(): 
    
    # init function to store car data.  
    def __init__(self):
        self.cars = []
        
    # Function for adding car in list
    def add_car(self, car):
        self.cars.append(car)   
        
    # Function for checking car is avaliable or not.   
    def aval_vehicals(self):
        avaliable_vehical=[car for car in self.cars if car.avaliable]
        if avaliable_vehical:
            print("Avaliable Cars")
            for car in avaliable_vehical:
                print(f' - {car}')
        else:
            print("No car avaliable")
    
    #Function to rent a vehical.
    def rent_vehical(self, car_id, days):
        for car in self.cars:
            if car.car_id==car_id and car.avaliable:
                rent_cost=car.rent_rate*days
                print(f"Rented {car} for {days} days. Total cost: ${rent_cost}")
                car.available = False
                return   
        print(f"Car with ID {car_id} not found or not available for rent.")    
    
    #Function to return cars.            
    def return_car(self, car_id):
        for car in self.cars:
            if car.car_id == car_id and not car.available:
                car.available = True
                print(f"Returned {car}.")
                return
        print(f"Car with ID {car_id} not found or not currently rented.")
       
# Decaling a variable for rent_system class.               
Car_system=rent_system()  

#Passing car details to the vehical class.     
car1 = Vehical(1, "Toyota", "Camry",  500)
car2 = Vehical(2, "Honda", "Accord", 650)
car3 = Vehical(3, "Ford", "Mustang", 850)    

#Passing car details to add_car function to append in list.
Car_system.add_car(car1)    
Car_system.add_car(car2) 
Car_system.add_car(car3)   

#Testing for renting different cars and returning cars.    
Car_system.aval_vehicals()
Car_system.rent_vehical(3,5)
Car_system.aval_vehicals()
Car_system.rent_vehical(2,3)
Car_system.return_car(3)

```

```python
#Problem 6:

#Declaring class Rectangle for find perimeter, area and volume of rectangle.
class Rectangle():
    
    # init function to take user input.
    def __init__(self,width, length, height):
        self.width=width
        self.length=length
        self.height=height
    
    # Function to calculate volume    
    def volume(self):
        return f'Volume of rectangle: {self.width*self.length*self.height} Cubic Units.'
    
    # Function to calculate area
    def area(self):
        return f'Area of rectangle: {self.width*self.length} Sq. Units.'
    
    # Function to calculate perimeter
    def perimeter(self):
        return f'Perimeter of rectangle: {2*(self.width+self.length)} Units.'

#Declaring class Square for find perimeter, area and volume of square.   
class Square():
    
    def __init__(self,side):
        self.side=side
    
    # Function to calculate volume
    def volume(self):
        return f'Volume of square: {self.side**3} Cubic Units.'
    
    # Function to calculate area
    def area(self):
        return f'Area of square: {self.side**2} Sq. Units.'
    
    # Function to calculate perimeter
    def perimeter(self):
        return f'Perimeter of square: {self.side*4} Units.'
 
#calculating volume of rectangle.  
find=Rectangle(5,3,7)
find.volume()

```
