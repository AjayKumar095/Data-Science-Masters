<h3>15. What is broadcasting in Numpy?</h3>
<p><b>Answer.</b></p>
<h4>Understanding Broadcasting in NumPy</h4>

<p>In NumPy, broadcasting is a powerful feature that allows arrays of different shapes and sizes to be combined, or "broadcasted," together in element-wise operations. This capability avoids the need for explicit looping over elements and makes it possible to perform operations on arrays of different shapes without creating additional copies of the data. Broadcasting rules define how these element-wise operations are carried out on arrays with different shapes.</p>

<p>The key idea behind broadcasting is to extend smaller arrays so that they have compatible shapes for arithmetic operations. The smaller array is "broadcast" across the larger array to make their shapes compatible. The broadcasting rules operate on the dimensions of the arrays, comparing them element-wise from the trailing dimensions backward.</p>

<p>Here are the main broadcasting rules:</p>

<ol>
  <li><strong>If the arrays do not have the same rank (number of dimensions):</strong>
    <ul>
      <li>Prepend dimensions of size 1 to the smaller array's shape until both shapes have the same length.</li>
    </ul>
  </li>
  <li><strong>For each dimension where the sizes differ:</strong>
    <ul>
      <li>If one of the sizes is 1, the array with size 1 is stretched to match the size of the other array.</li>
    </ul>
  </li>
  <li><strong>If the sizes along a dimension are neither 1 nor equal:</strong>
    <ul>
      <li>Broadcasting is not possible, and NumPy will raise a <code>ValueError</code>.</li>
    </ul>
  </li>
</ol>

<p>Here's a simple example to illustrate broadcasting:</p>

<pre>
<code>
import numpy as np
a = np.array([1.0, 2.0, 3.0])
b = 2.0

result = a * b
print("Array 'a':", a)
print("Scalar 'b':", b)
print("Result after broadcasting:", result)
</code>
</pre>

<p>In this example, the scalar <code>b</code> is broadcast to the array <code>a</code> during the multiplication operation. The scalar is effectively treated as a 1D array with the same shape as <code>a</code>. The result is a new array where each element of <code>a</code> is multiplied by the scalar <code>b</code>.</p>

<p>Broadcasting simplifies code and makes it more readable, eliminating the need for explicit looping or reshaping of arrays in many cases. It is a fundamental concept in NumPy that facilitates efficient element-wise operations on arrays of different shapes.</p>
