```python
# Problem 1:

# declaring a list or array with duplicates values.
dupli_list=[1,2,3,1,2,4,2,2]

# declaring the variables that store number frequency and number itself.
number_frequency=0 
element =None

# Finding the element or number which appear maximum time in array
for i in (dupli_list):
    count=0
    for j in dupli_list:
        if i==j:
          count+=1
    if count>number_frequency:
      number_frequency=count
      element=i
 
# Printing the result.     
print(f"Element {element} appear maximum time in the list having appearing frequency of {number_frequency}.")
```

```python
# Problem 2:

# Let assume we have a list or array in a range of 1 to n.
# Where n be any interger value.
# Also there was no missing values. And only one value is missing.

array_with_missing_value=[1,2,4,5,6,3,7,8,9] # This is array from range(1,10), having a missing value which is "5" 

# let sort the list in assending order as it is a continous array with difference of 1 between to elements.
# And find the missing value.
def sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place, so we don't need to check them
        for j in range(0, n-i-1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
               arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
# Here we find the missing element in the given array
def finding_missing(sorted_arr):
    for i in range(len(sorted_arr) - 1):
        if sorted_arr[i + 1] - sorted_arr[i] != 1:
            return sorted_arr[i]+1
        if i + 1 == len(sorted_arr) - 1:
           break 
    return None   

# passing the array to the function.       
arr=sort(array_with_missing_value)

# Result.
missing_num=finding_missing(arr)
if missing_num!=None:
    print("Missing element is",missing_num)
else:
    print("No missing element in given range.")

# Note: - This algorithum is works, assuming the list or array element is once sorted and have difference between two element is one.
# And array start from 1 to n numbers.
```
