```python
"""12. Explain the use of the `super()` function in Python constructors. Provide an example."""

"""Answer.
    In Python, the super() function is used in constructors to call the constructor of the parent (or superclass) class.
    It is particularly useful in situations where you have overridden the constructor in a subclass"""
    
# Exapmle
class ParentClass:
    def __init__(self, parent_attribute):
        print("Inside ParentClass constructor")
        self.parent_attribute = parent_attribute

class ChildClass(ParentClass):
    def __init__(self, parent_attribute, child_attribute):
        # Calling the constructor of the parent class using super()
        super().__init__(parent_attribute)
        
        # Initializing attributes specific to the child class
        print("Inside ChildClass constructor")
        self.child_attribute = child_attribute

# Creating an instance of ChildClass
child_object = ChildClass("Parent Value", "Child Value")

# Accessing attributes of both parent and child classes
print("Parent Attribute:", child_object.parent_attribute)
print("Child Attribute:", child_object.child_attribute)
```

```python
"""19. Explain the use of constructor chaining in Python. Provide a practical example."""

"""Answer.
    
Constructor chaining in Python refers to the concept of one constructor calling
another constructor within the same class or in its parent class.
This allows for reusing the initialization logic and avoiding code duplication."""

# Example
class ParentClass:
    def __init__(self, parent_attribute):
        print("Inside ParentClass constructor")
        self.parent_attribute = parent_attribute

class ChildClass(ParentClass):
    def __init__(self, parent_attribute, child_attribute):
        # Call the constructor of the parent class using super()
        super().__init__(parent_attribute)

        print("Inside ChildClass constructor")
        self.child_attribute = child_attribute


# Creating childclass instance 
child_object = ChildClass("Parent Value", "Child Value")

# printing result
print("Parent Attribute:", child_object.parent_attribute)
print("Child Attribute:", child_object.child_attribute)
```

```python
"""5. How can you access the methods and attributes of a parent class from a child class in Python? Give an
example."""

"""Answer
In Python, you can access the methods and attributes of a parent class from a child class using the super() function."""

# Example
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    def __init__(self, species, breed):
    
        super().__init__(species)
        self.breed = breed

    def make_sound(self):

        parent_sound = super().make_sound()
        return f"{parent_sound} - Woof! Woof!"

# Creating an instance 
my_dog = Dog(species="Canine", breed="Golden Retriever")

# Accessing attributes of the parent class 
print(f"Species: {my_dog.species}")
print(f"Breed: {my_dog.breed}")

# Calling the method 
print(my_dog.make_sound())
```

```python
"""8. Explain the role of the `isinstance()` function in Python and how it relates to inheritance."""
"""Answer
        The isinstance() function in Python is used to check if an object belongs to a specified class or a tuple of classes.
        It returns True if the object is an instance of any of the specified classes and False otherwise."""

# Example
class Animal:
    def speak(self):
        return "Generic animal sound"

class Dog(Animal):
    def bark(self):
        return "Woof! Woof!"

class Cat(Animal):
    def meow(self):
        return "Meow! Meow!"

# Creating instances 
animal_instance = Animal()
dog_instance = Dog()
cat_instance = Cat()

# Using isinstance() 
print(isinstance(animal_instance, Animal))  
print(isinstance(dog_instance, Animal))     
print(isinstance(cat_instance, Animal))    

print(isinstance(dog_instance, Dog))     
print(isinstance(cat_instance, Dog))      

print(isinstance(animal_instance, (Dog, Cat))) 
        
```

<h2>18. What is the "diamond problem" in multiple inheritance, and how does Python address it?</h2>
<p><b>Answer.</b></p>
<h4>The "Diamond Problem" in Multiple Inheritance</h4>

<p><strong>Definition:</strong></p>
    <ul>
        <li>The "diamond problem" is a term used in object-oriented programming to describe an ambiguity that arises when a class inherits from two classes that have a common ancestor.</li>
        <li>It occurs in languages that support multiple inheritance, where a class can inherit from more than one parent class.</li>
        <li>The problem is named "diamond" because of the shape of the inheritance diagram.</li>
    </ul>

<p><strong>Scenario:</strong></p>
    <ul>
        <li>Consider classes A and B, both of which inherit from a common class, C.</li>
        <li>Now, class D inherits from both A and B.</li>
        <li>If there is a method or attribute in C that is overridden in both A and B, it creates ambiguity for class D because it inherits from both A and B.</li>
        <li>The question arises: which overridden method or attribute should be used in class D?</li>
    </ul>

<p><strong>How Python Addresses the Diamond Problem:</strong></p>
    <ul>
        <li>Python uses a technique called "C3 linearization" to determine the order in which classes are considered during method resolution in multiple inheritance.</li>
        <li>It ensures a consistent and predictable order of method resolution, avoiding ambiguity.</li>
        <li>The order is determined based on the order of class declaration and the order in which base classes are listed in the class definition.</li>
        <li>This order ensures that the most specific method in the hierarchy is chosen, addressing the diamond problem.</li>
    </ul>

 <h4>Concepts of "Is-a" and "Has-a" Relationships in Inheritance</h4>

<p><strong>"Is-a" Relationship (Inheritance):</strong></p>
    <ul>
        <li>An "is-a" relationship represents a hierarchy where one class is a specialized version of another class.</li>
        <li>It is often implemented through inheritance, allowing a subclass to inherit attributes and behaviors from a superclass.</li>
        <li>This relationship implies that the subclass "is a" type of the superclass.</li>
    </ul>

<pre>
<code>
class Animal:
    def speak(self):
        return "Generic animal sound"

class Dog(Animal):
    def bark(self):
        return "Woof!"

print(my_dog.bark())   
</code>
    </pre>

<p><strong>"Has-a" Relationship (Composition):</strong></p>
    <ul>
        <li>A "has-a" relationship represents a composition where one class contains an instance of another class as one of its attributes.</li>
        <li>It is often implemented through composition, allowing a class to have an object of another class as part of its structure.</li>
        <li>This relationship implies that the class "has" another class as a component.</li>

    </pre>
