```python
"""12. Explain the use of the `super()` function in Python constructors. Provide an example."""

"""Answer.
    In Python, the super() function is used in constructors to call the constructor of the parent (or superclass) class.
    It is particularly useful in situations where you have overridden the constructor in a subclass"""
    
# Exapmle
class ParentClass:
    def __init__(self, parent_attribute):
        print("Inside ParentClass constructor")
        self.parent_attribute = parent_attribute

class ChildClass(ParentClass):
    def __init__(self, parent_attribute, child_attribute):
        # Calling the constructor of the parent class using super()
        super().__init__(parent_attribute)
        
        # Initializing attributes specific to the child class
        print("Inside ChildClass constructor")
        self.child_attribute = child_attribute

# Creating an instance of ChildClass
child_object = ChildClass("Parent Value", "Child Value")

# Accessing attributes of both parent and child classes
print("Parent Attribute:", child_object.parent_attribute)
print("Child Attribute:", child_object.child_attribute)
```

```python
"""19. Explain the use of constructor chaining in Python. Provide a practical example."""

"""Answer.
    
Constructor chaining in Python refers to the concept of one constructor calling
another constructor within the same class or in its parent class.
This allows for reusing the initialization logic and avoiding code duplication."""

# Example
class ParentClass:
    def __init__(self, parent_attribute):
        print("Inside ParentClass constructor")
        self.parent_attribute = parent_attribute

class ChildClass(ParentClass):
    def __init__(self, parent_attribute, child_attribute):
        # Call the constructor of the parent class using super()
        super().__init__(parent_attribute)

        print("Inside ChildClass constructor")
        self.child_attribute = child_attribute


# Creating childclass instance 
child_object = ChildClass("Parent Value", "Child Value")

# printing result
print("Parent Attribute:", child_object.parent_attribute)
print("Child Attribute:", child_object.child_attribute)
```

```python
"""5. How can you access the methods and attributes of a parent class from a child class in Python? Give an
example."""

"""Answer
In Python, you can access the methods and attributes of a parent class from a child class using the super() function."""

# Example
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    def __init__(self, species, breed):
    
        super().__init__(species)
        self.breed = breed

    def make_sound(self):

        parent_sound = super().make_sound()
        return f"{parent_sound} - Woof! Woof!"

# Creating an instance 
my_dog = Dog(species="Canine", breed="Golden Retriever")

# Accessing attributes of the parent class 
print(f"Species: {my_dog.species}")
print(f"Breed: {my_dog.breed}")

# Calling the method 
print(my_dog.make_sound())
```

```python
"""8. Explain the role of the `isinstance()` function in Python and how it relates to inheritance."""
"""Answer
        The isinstance() function in Python is used to check if an object belongs to a specified class or a tuple of classes.
        It returns True if the object is an instance of any of the specified classes and False otherwise."""

# Example
class Animal:
    def speak(self):
        return "Generic animal sound"

class Dog(Animal):
    def bark(self):
        return "Woof! Woof!"

class Cat(Animal):
    def meow(self):
        return "Meow! Meow!"

# Creating instances 
animal_instance = Animal()
dog_instance = Dog()
cat_instance = Cat()

# Using isinstance() 
print(isinstance(animal_instance, Animal))  
print(isinstance(dog_instance, Animal))     
print(isinstance(cat_instance, Animal))    

print(isinstance(dog_instance, Dog))     
print(isinstance(cat_instance, Dog))      

print(isinstance(animal_instance, (Dog, Cat))) 
        
```

<h4>The "Diamond Problem" in Multiple Inheritance</h4>

<p><strong>Definition:</strong></p>
    <ul>
        <li>The "diamond problem" is a term used in object-oriented programming to describe an ambiguity that arises when a class inherits from two classes that have a common ancestor.</li>
        <li>It occurs in languages that support multiple inheritance, where a class can inherit from more than one parent class.</li>
        <li>The problem is named "diamond" because of the shape of the inheritance diagram.</li>
    </ul>

<p><strong>Scenario:</strong></p>
    <ul>
        <li>Consider classes A and B, both of which inherit from a common class, C.</li>
        <li>Now, class D inherits from both A and B.</li>
        <li>If there is a method or attribute in C that is overridden in both A and B, it creates ambiguity for class D because it inherits from both A and B.</li>
        <li>The question arises: which overridden method or attribute should be used in class D?</li>
    </ul>

<p><strong>How Python Addresses the Diamond Problem:</strong></p>
    <ul>
        <li>Python uses a technique called "C3 linearization" to determine the order in which classes are considered during method resolution in multiple inheritance.</li>
        <li>It ensures a consistent and predictable order of method resolution, avoiding ambiguity.</li>
        <li>The order is determined based on the order of class declaration and the order in which base classes are listed in the class definition.</li>
        <li>This order ensures that the most specific method in the hierarchy is chosen, addressing the diamond problem.</li>
    </ul>

```python
"""12. Explain the concept of property decorators in Python and how they relate to encapsulation."""

"""Answer
In Python, property decorators are a way to implement getter, setter, and deleter methods for class attributes.
They provide a convenient syntax for accessing and modifying the values of attributes 
while allowing the developer to control the behavior of these operations."""

class MyClass:
    def __init__(self):
        self._value = 0  # protected attribute

    @property
    def value(self):
        return self._value  # getter method

    @value.setter
    def value(self, new_value):
        if new_value >= 0:
            self._value = new_value  # setter method
        else:
            print("Invalid value")

    @value.deleter
    def value(self):
        print("Deleting value")
        del self._value  # deleter method

# Example usage
obj = MyClass()

# Accessing value using the property
current_value = obj.value
print("Current Value:", current_value)

# Setting value using the property
obj.value = 42

# Deleting value using the property
del obj.value
```

```python
"""13. What is data hiding, and why is it important in encapsulation? Provide examples."""

"""Answer
        
Data hiding is a concept in object-oriented programming (OOP) that involves restricting access to the internal details of an object.
In the context of encapsulation, data hiding is achieved by making the implementation details,
such as attributes and methods, private or protected, meaning they are not directly accessible from outside the class."""

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def get_balance(self):
        return self.__balance  # getter method for private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

# Example usage
account = BankAccount(1000)

# Accessing balance using the getter method
current_balance = account.get_balance()
print("Current Balance:", current_balance)

# Depositing and withdrawing funds
account.deposit(500)
account.withdraw(200)

# Attempting direct access to the private attribute (not recommended)
# This will raise an AttributeError due to data hiding.
try:
    print(account.__balance)
except AttributeError as e:
    print(f"AttributeError: {e}")
```

```python
"""17. Create a Python class for a library system that encapsulates book information, including titles, authors,
and availability status."""

class Book:
    def __init__(self, title, author, availability=True):
        self.__title = title            # private attribute
        self.__author = author          # private attribute
        self.__availability = availability  # private attribute

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def is_available(self):
        return self.__availability

    def borrow_book(self):
        if self.__availability:
            print(f"The book '{self.__title}' by {self.__author} has been borrowed.")
            self.__availability = False
        else:
            print(f"The book '{self.__title}' is currently not available.")

    def return_book(self):
        if not self.__availability:
            print(f"The book '{self.__title}' has been returned.")
            self.__availability = True
        else:
            print(f"The book '{self.__title}' is already available.")

# Example usage
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")

# Displaying book information
print("Book Information:")
print(f"Title: {book1.get_title()}, Author: {book1.get_author()}, Available: {book1.is_available()}")
print(f"Title: {book2.get_title()}, Author: {book2.get_author()}, Available: {book2.is_available()}")

# Borrowing and returning books
book1.borrow_book()
book2.borrow_book()

book1.return_book()
book2.return_book()

# Displaying updated book availability
print("\nUpdated Book Availability:")
print(f"Title: {book1.get_title()}, Available: {book1.is_available()}")
print(f"Title: {book2.get_title()}, Available: {book2.is_available()}")
```

<h3>19. Describe the concept of information hiding in encapsulation. Why is it essential in software development?</h3>
<p><b>Answer.</b></p>
<p>
  In software development, the concept of <strong>information hiding</strong> is an integral aspect of encapsulation. Information hiding involves the deliberate concealment of implementation details within a class, exposing only a well-defined public interface to external code. This is achieved through the use of access modifiers like private and protected, restricting direct access to certain attributes or methods.
</p>

<ul>
  <li><strong>Enhanced Security:</strong> Information hiding contributes to enhanced security by preventing unauthorized access to sensitive data and implementation details. Private attributes or methods can only be accessed or modified through controlled interfaces, reducing the risk of data misuse or corruption.</li>

  <li><strong>Reduced Complexity:</strong> By hiding the internal workings of a class, information hiding reduces the complexity of external code. Developers interact with the public interface, focusing on the essential functionalities without being burdened by intricate implementation details.</li>

  <li><strong>Improved Maintainability:</strong> Concealing implementation details allows for better maintainability. Changes to the internal structure of a class, such as modifying attributes or methods, do not impact external code as long as the public interface remains unchanged. This isolation simplifies updates and enhancements.</li>

  <li><strong>Facilitates Evolution:</strong> Information hiding facilitates the evolution of software systems. Developers can evolve the internal implementation of a class without affecting the overall functionality. This separation allows for the introduction of new features or improvements without disrupting existing code.</li>
</ul>
