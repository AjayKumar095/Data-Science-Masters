```python
""""1. Given an array, check if it contains any duplicates or not."""

#Function to check if array contains any duplicate or not.
def check_dupli(arr):
    #Declaring a empty set.
    check=set()

    #For loop to itrate through array of elements.
    for i in arr:
        if i in check:
            return True
        check.add(i)
    return False

#Example array.
Array=[1, 2, 4, 2, 5, 9]

#Function call
Check_my_array=check_dupli(Array)

#Printing result
if Check_my_array:
    print(f'{Check_my_array}')
else:
    print(f'{Check_my_array}')
```

<h4>Iterating Through the Array:</h4>

<h4>Checking for Duplicates:</h4>
<p>For each element, it checks if the element (<code>i</code>) is already present in the set (<code>check</code>).</p>
<p>If the element is found in the set, it means a duplicate exists, and the function returns <code>True</code>.</p>

<h4>Adding Elements to the Set:</h4>
<p>If the element is not found in the set, it is added to the set using the <code>add</code> method.</p>

<h4>Returning Result:</h4>

```python
"""2. Given an array and an integer k, rotate the array to the right by k steps."""


#Function to rotate the array.
def rotate(nums, k):
    # Calculate the effective rotation amount to handle cases where k is larger than the array length
    k = k % len(nums)
    
    # Reverse the entire array
    reverse(nums, 0, len(nums) - 1)
    
    # Reverse the first k elements
    reverse(nums, 0, k - 1)
    
    # Reverse the remaining elements
    reverse(nums, k, len(nums) - 1)

#Function to reverse the array.
def reverse(nums, start, end):
    while start < end:
        nums[start], nums[end] = nums[end], nums[start]
        start += 1
        end -= 1

#input arry
my_array = [1, 2, 3, 4, 5, 6, 7]
k = 3
rotate(my_array, k)
print("Output =",my_array)
```

<h1><center>Explanation</center></h1>

<h4>Rotate Function:</h4>
<p>
  The rotate function takes two parameters: <code>nums</code> (the array to be rotated) and <code>k</code> (the number of steps to rotate the array to the right).<br>
  It first calculates the effective rotation amount by taking the remainder of <code>k</code> divided by the length of the array. This ensures that excessive rotations are handled.
</p>

<h4>Reverse Function:</h4>
<p>
  The <code>reverse</code> function takes three parameters: <code>nums</code> (the array), <code>start</code> (the starting index), and <code>end</code> (the ending index).<br>
  It reverses the elements in the specified range [<code>start</code>, <code>end</code>] using a two-pointer approach.
</p>

<h4>Reverse the Entire Array:</h4>
<p>
  The <code>reverse</code> function is called to reverse the entire array (<code>nums</code>) at the beginning of the <code>rotate</code> function.
</p>

<h4>Reverse the First k Elements:</h4>
<p>
  The <code>reverse</code> function is called again to reverse the first <code>k</code> elements of the array (<code>nums[0:k-1]</code>).
</p>

<h4>Reverse the Remaining Elements:</h4>
<p>
  Once again, the <code>reverse</code> function is called to reverse the remaining elements in the array (<code>nums[k:len(nums)-1]</code>).
</p>

<h4>Input Array and Rotation:</h4>
<p>
  An input array (<code>my_array = [1, 2, 3, 4, 5, 6, 7]</code>) is provided, and the number of steps to rotate the array to the right is set to <code>k = 3</code>.
</p>

<h4>Rotate the Input Array:</h4>
<p>
  The <code>rotate</code> function is called with the input array and the specified number of steps.
</p>

<h4>Print the Result:</h4>
<p>
  The rotated array is printed as output.
</p>

<h4>Example Output:</h4>
<p>
  If the input array is <code>[1, 2, 3, 4, 5, 6, 7]</code> and <code>k</code> is set to 3, the output would be the rotated array: <code>[5, 6, 7, 1, 2, 3, 4]</code>.
</p>




```python
"""3. Reverse the given array in-place, means without using any extra data structure."""

# Function to reverse the array.
def reverse(arr):
    start_index = 0
    end_index = len(arr) - 1

    while start_index < end_index:
        # Swap elements at start_index and end_index
        arr[start_index], arr[end_index] = arr[end_index], arr[start_index]

        # Move indices towards the center
        start_index += 1
        end_index -= 1
    return arr

#Input array.
my_array = [2, 4, 5, 7, 9, 12]

#passing array to the function.
result=reverse(my_array)

#Printing reversed array.
print("Output =", result)
```

<h1><center>Explanation</center></h1>

<h4>Function Definition:</h4>
    <p>The program defines a function named <code>reverse</code> that takes an array (<code>arr</code>) as its parameter.</p>

<h4>Initialize Indices:</h4>
    <ul>
        <li>Two indices, <code>start_index</code> and <code>end_index</code>, are initialized at the beginning and end of the array, respectively.</li>
        <li><code>start_index</code> is set to 0, and <code>end_index</code> is set to <code>len(arr) - 1</code>.</li>
    </ul>

<h4>Swap Elements in a Loop:</h4>
    <ul>
        <li>The program enters a <code>while</code> loop, which continues as long as <code>start_index</code> is less than <code>end_index</code>.</li>
        <li>Inside the loop, it swaps the elements at <code>start_index</code> and <code>end_index</code>.</li>
    </ul>

<h4>Move Indices Towards the Center:</h4>
    <ul>
        <li>After the swap, both indices are moved towards the center of the array.</li>
        <li><code>start_index</code> is incremented by 1, and <code>end_index</code> is decremented by 1.</li>
    </ul>

<h4>Return Reversed Array:</h4>
    <p>The <code>reverse</code> function returns the reversed array.</p>

<h4>Input Array:</h4>
    <p>An example array <code>my_array</code> is defined with values [2, 4, 5, 7, 9, 12].</p>

<h4>Function Invocation:</h4>
    <p>The <code>reverse</code> function is called with <code>my_array</code> as an argument, and the result is stored in the variable <code>result</code>.</p>

<h4>Print Reversed Array:</h4>
    <p>The program prints the reversed array using the <code>print</code> statement.</p>

<h4>Example Output:</h4>
    <p>If the input array is [2, 4, 5, 7, 9, 12], the output will be "Output = [12, 9, 7, 5, 4, 2]".</p>

 


```python
"""4. Given an array of integers, find the maximum element in an array"""

#Declaring a function for finding maximum element is arry.
def max_of(arr):
    max_value=arr[0]
    for num in arr:
      if num > max_value:
        max_value = num
    return max_value

    
arr = [10, 5, 20, 8, 15]
Maximum_element=max_of(arr)
print(f"Output = {Maximum_element}") 
```

```python
""""5. Given a sorted array, remove the duplicate element without using any extra data structure."""

#Function to check if array contains any duplicate or not.
def check_dupli(arr):
    #Declaring a empty set.
    check=set()
     
    #For loop to itrate through array of elements.
    for i in arr:
        if i in check:
            continue
        else:
            check.add(i)
    return check    
        
   
#Declaring a array of duplicate elements.
Array=[1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]

#Printing the final result.
Check_my_array=check_dupli(Array)
output=[]
for i in Check_my_array:
    output.append(i)
print("Output=",output)
```
